```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.show

plt.plot(x,y, label='x^2')
plt.title('Simple Line Plot')
plt.xlabel('x Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
categories = ['A','B','C']
values = [7,3,5]
plt.bar(categories, values)
plt.title('Bar Chart')
plt.xlabel('categories')
plt.ylabel('values')
plt.show
```
